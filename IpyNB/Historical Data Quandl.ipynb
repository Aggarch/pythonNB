```python
#create a merge dataframe OHLC resample 10 min
#plot dataframe
```

```python
data_frame.Ask=data_frame.Ask/100.0
data_frame.Bid=data_frame.Bid/100.0
df_bid=data_frame[['Bid','BidSize']]
df_ask=data_frame[['Ask','AskSize']]
```

```python
ohlc_dict = {'open':'first','high':'max','low':'min','close':'last', 'volume':'sum'  }
df_bid_ohlc=df_bid['Bid'].resample('10min').ohlc()
df_bid_ohlc['volume']=df_bid['BidSize'].resample('10min').sum()
df_ask_ohlc=df_ask['Ask'].resample('10min').ohlc()
df_ask_ohlc['volume']=df_ask['AskSize'].resample('10min').sum()
```

```python
df_ask_bid=pd.concat([df_bid_ohlc, df_ask_ohlc], axis=1, keys=['Bid', 'Ask'])
```

```python
%matplotlib inline
df_ask_bid.Bid.close.plot()
```
