```python
import pandas as pd
import os
```

```python
default_path="C:\\Users\\junaj\\Desktop\\CursoDataScience\\notebooks\\"
data_path="./data/"
os.chdir(default_path)
```

```python
#Load fxs the data https://pepperstone.com/uk/client-resources/historical-tick-data
data_frame = pd.read_csv(data_path+'EURUSD-2016-01.csv', names=['Symbol', 'Date_Time', 'Bid', 'Ask'], index_col=1, parse_dates=True)
```

```python
#data frame looks like:
data_frame.head(10)
```

```python
#resample attribute of pandas data frame
```

```python
timeframe='15Min'
data_ask =  data_frame['Ask'].resample(timeframe).ohlc()

data_bid =  data_frame['Bid'].resample(timeframe).ohlc()
```

```python
data_bid.head(5)
```

```python
data_ask.head(5)
```

```python
df_ask_bid=pd.concat([data_bid, data_ask], axis=1, keys=['Bid', 'Ask'])
```
